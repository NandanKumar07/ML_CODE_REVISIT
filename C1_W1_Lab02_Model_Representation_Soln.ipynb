```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
x_train = np.array([1.0, 2.0])  # size in 1000 sqft
y_train = np.array([300.0, 500.0])   # price in 1000 of dollars
```

```python
print(f"x_train : {x_train}")
print(f"y_train : {y_train}")
```

```python
# No. of Training example

m = x_train.shape[0]
m
```

```python
# plotting the data
plt.scatter(x_train, y_train, marker = 'x', color = 'r')
plt.title("Housing prices")
plt.xlabel("size in 1000 sqft")
plt.ylabel("price in 1000 Dollars")
plt.show()
```

```python
# let w = 100, b = 100
w = 200
b = 100
```

```python
def compute_model_output(x, w, b):
    m = x.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i] + b
    
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)
```

```python
tmp_f_wb
```

```python
plt.plot(x_train, tmp_f_wb, c = 'b', label = 'Our Prediction')

plt.scatter(x_train, y_train, marker = 'x', c = 'r', label = 'Actual Values')

plt.title("Housing prices")
plt.xlabel("size in 1000 sqft")
plt.ylabel("price in 1000 Dollars")
plt.legend()
plt.show()
```

```python
# Now for predicting the values, we have find the most accurate w and b

x_i = float(input())
cost = w*x_i + b
print(f"Price for {x_i*1000} sqft area is: ${cost} thousand dollars")
```
